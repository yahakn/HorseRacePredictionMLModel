```python
import pandas as pd
import os
import chardet

folder_path = "/content/drive/My Drive/koşular"

header = ["At No", "At İsmi", "Yaş", "Orijin(Baba)", "Orijin(Anne)", "Kilo",
          "Jokey Adı", "Sahip Adı", "Antrenör Adı", "St", "AGF", "H", "Derece", "Ganyan", "Fark"]

all_data = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        # Dosya encoding’ini tespit et
        with open(file_path, 'rb') as f:
            raw_data = f.read(10000)
            result = chardet.detect(raw_data)
            encoding = result['encoding']

        with open(file_path, 'r', encoding=encoding, errors='ignore') as f:
            lines = f.readlines()

        data = []
        found_header = False

        for line in lines:
            line = line.strip()
            if found_header:
                if line == "" or "Kosu" in line:
                    found_header = False
                    continue
                row = line.split(";")
                if len(row) == len(header):
                    data.append(row)
            if line.replace(" ", "").startswith("AtNo;Atİsmi;"):
                found_header = True

        all_data.extend(data)

df = pd.DataFrame(all_data, columns=header)
print(df.head())
```

```python
df = df.drop_duplicates()
```

```python
dosya_yolu = "/content/drive/My Drive/Adsız klasör/tarih/birlesik_df.csv"
import pandas as pd
tarih = pd.read_csv(dosya_yolu)
tarih.head()  # Veriyi görüntüle
```

```python
tarih = tarih.drop_duplicates()
```

```python
tarih = tarih.rename(columns={
    "Birinci": "At İsmi",
    "H. Puanı": "H"
})
```

```python
for d in [df, tarih]:
    d["At İsmi"] = d["At İsmi"].astype(str).str.strip().str.upper()
    d["Derece"] = d["Derece"].astype(str).str.strip().str.replace(":", ".", regex=False)

    # H sütununu sayıya çevirip sonra string yap
    d["H"] = pd.to_numeric(d["H"], errors="coerce").fillna(0).astype(int).astype(str)
```

```python
# Küçük-büyük harf uyumsuzluklarına karşı normalize edelim
df["Derece"] = df["Derece"].str.strip().str.lower()

# "derecesiz" ve "koşmaz" içeren satırları filtrele
mask = df["Derece"].isin(["derecesiz", "koşmaz"])
print(f"Silinecek satır sayısı: {mask.sum()}")

# Bu satırları veri setinden çıkaralım
df = df[~mask].copy()
```

```python
import pandas as pd
data = df
# 📌 DataFrame oluşturuyoruz
df = pd.DataFrame(data)

# 📌 At ismiyle ilgili olmayan kısaltmalar listesi
kisaltmalar = ["KG", "KUL", "DB", "BB", "ÇABA", "GKR", "SKG", "SKUL", "SGKR", "YP", "TGK", "ÖG", "DS", "SK", "K"]

# 📌 At ismi ve detayları ayıran fonksiyon (while ile)
def ayir_at_isim(at_isim):
    kelimeler = at_isim.split()
    detaylar = []

    # Sondan başlayarak, kısaltmaları ayır
    while kelimeler and kelimeler[-1] in kisaltmalar:
        detaylar.append(kelimeler.pop())  # Son kelimeyi detaylara ekle

    return " ".join(kelimeler), " ".join(reversed(detaylar)) if detaylar else None

# 📌 Yeni sütunları ekleyelim
df[["At İsmi", "Detaylar"]] = df["At İsmi"].apply(lambda x: pd.Series(ayir_at_isim(x)))

# 📌 Sonuçları görelim
df
```

```python
# At No == 1 olan satırları işaretle
yeni_yaris_baslangici = (df["At No"].astype(str).str.strip() == "1").astype(int)

# Cumsum ile yarış bloklarını sırala
df["Yarış_Id"] = yeni_yaris_baslangici.cumsum()
```

```python
df_1 = df[df["At No"] == "1"]
```

```python
tarih = tarih.dropna(subset=["Şehir"])
```

```python
for d in [df_1, tarih]:
    d["At İsmi"] = d["At İsmi"].astype(str).str.strip().str.upper()
    d["Derece"] = d["Derece"].astype(str).str.strip().str.lower()
    d["H"] = d["H"].astype(str).str.strip().replace("", "Unknown").fillna("Unknown")
```

```python
tarih_aktar = tarih[["At İsmi", "Derece", "H", "Tarih", "Şehir", "Pist", "Mesafe", "Sıklet", "Koşu"]].drop_duplicates()
```

```python
df_1 = df_1.merge(
    tarih_aktar,
    on=["At İsmi", "Derece", "H"],
    how="left"
)
```
