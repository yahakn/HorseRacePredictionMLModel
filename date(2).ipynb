```python
import pandas as pd
import os
import chardet

folder_path = "/content/drive/My Drive/koÅŸular"

header = ["At No", "At Ä°smi", "YaÅŸ", "Orijin(Baba)", "Orijin(Anne)", "Kilo",
          "Jokey AdÄ±", "Sahip AdÄ±", "AntrenÃ¶r AdÄ±", "St", "AGF", "H", "Derece", "Ganyan", "Fark"]

all_data = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        # Dosya encodingâ€™ini tespit et
        with open(file_path, 'rb') as f:
            raw_data = f.read(10000)
            result = chardet.detect(raw_data)
            encoding = result['encoding']

        with open(file_path, 'r', encoding=encoding, errors='ignore') as f:
            lines = f.readlines()

        data = []
        found_header = False

        for line in lines:
            line = line.strip()
            if found_header:
                if line == "" or "Kosu" in line:
                    found_header = False
                    continue
                row = line.split(";")
                if len(row) == len(header):
                    data.append(row)
            if line.replace(" ", "").startswith("AtNo;AtÄ°smi;"):
                found_header = True

        all_data.extend(data)

df = pd.DataFrame(all_data, columns=header)
print(df.head())
```

```python
df = df.drop_duplicates()
```

```python
for d in [df, tarih]:
    d["At Ä°smi"] = d["At Ä°smi"].astype(str).str.strip().str.upper()
    d["Derece"] = d["Derece"].astype(str).str.strip().str.replace(":", ".", regex=False)

    # H sÃ¼tununu sayÄ±ya Ã§evirip sonra string yap
    d["H"] = pd.to_numeric(d["H"], errors="coerce").fillna(0).astype(int).astype(str)
```

```python
# KÃ¼Ã§Ã¼k-bÃ¼yÃ¼k harf uyumsuzluklarÄ±na karÅŸÄ± normalize edelim
df["Derece"] = df["Derece"].str.strip().str.lower()

# "derecesiz" ve "koÅŸmaz" iÃ§eren satÄ±rlarÄ± filtrele
mask = df["Derece"].isin(["derecesiz", "koÅŸmaz"])
print(f"Silinecek satÄ±r sayÄ±sÄ±: {mask.sum()}")

# Bu satÄ±rlarÄ± veri setinden Ã§Ä±karalÄ±m
df = df[~mask].copy()
```

```python
import pandas as pd
data = df
# ðŸ“Œ DataFrame oluÅŸturuyoruz
df = pd.DataFrame(data)

# ðŸ“Œ At ismiyle ilgili olmayan kÄ±saltmalar listesi
kisaltmalar = ["KG", "KUL", "DB", "BB", "Ã‡ABA", "GKR", "SKG", "SKUL", "SGKR", "YP", "TGK", "Ã–G", "DS", "SK", "K"]

# ðŸ“Œ At ismi ve detaylarÄ± ayÄ±ran fonksiyon (while ile)
def ayir_at_isim(at_isim):
    kelimeler = at_isim.split()
    detaylar = []

    # Sondan baÅŸlayarak, kÄ±saltmalarÄ± ayÄ±r
    while kelimeler and kelimeler[-1] in kisaltmalar:
        detaylar.append(kelimeler.pop())  # Son kelimeyi detaylara ekle

    return " ".join(kelimeler), " ".join(reversed(detaylar)) if detaylar else None

# ðŸ“Œ Yeni sÃ¼tunlarÄ± ekleyelim
df[["At Ä°smi", "Detaylar"]] = df["At Ä°smi"].apply(lambda x: pd.Series(ayir_at_isim(x)))

# ðŸ“Œ SonuÃ§larÄ± gÃ¶relim
df
```

```python
# At No == 1 olan satÄ±rlarÄ± iÅŸaretle
yeni_yaris_baslangici = (df["At No"].astype(str).str.strip() == "1").astype(int)

# Cumsum ile yarÄ±ÅŸ bloklarÄ±nÄ± sÄ±rala
df["YarÄ±ÅŸ_Id"] = yeni_yaris_baslangici.cumsum()
```
