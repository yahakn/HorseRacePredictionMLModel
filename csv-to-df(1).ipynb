```python
import pandas as pd
import os
import chardet

folder_path = "/content/drive/My Drive/koşular"

header = ["At No", "At İsmi", "Yaş", "Orijin(Baba)", "Orijin(Anne)", "Kilo",
          "Jokey Adı", "Sahip Adı", "Antrenör Adı", "St", "AGF", "H", "Derece", "Ganyan", "Fark"]

all_data = []

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        # Dosya encoding’ini tespit et
        with open(file_path, 'rb') as f:
            raw_data = f.read(10000)
            result = chardet.detect(raw_data)
            encoding = result['encoding']

        with open(file_path, 'r', encoding=encoding, errors='ignore') as f:
            lines = f.readlines()

        data = []
        found_header = False

        for line in lines:
            line = line.strip()
            if found_header:
                if line == "" or "Kosu" in line:
                    found_header = False
                    continue
                row = line.split(";")
                if len(row) == len(header):
                    data.append(row)
            if line.replace(" ", "").startswith("AtNo;Atİsmi;"):
                found_header = True

        all_data.extend(data)

df = pd.DataFrame(all_data, columns=header)
print(df.head())
```

```python
import pandas as pd
data = df
# 📌 DataFrame oluşturuyoruz
df = pd.DataFrame(data)

# 📌 At ismiyle ilgili olmayan kısaltmalar listesi
kisaltmalar = ["KG", "KUL", "DB", "BB", "ÇABA", "GKR", "SKG", "SKUL", "SGKR", "YP", "TGK", "ÖG", "DS", "SK", "K"]

# 📌 At ismi ve detayları ayıran fonksiyon (while ile)
def ayir_at_isim(at_isim):
    kelimeler = at_isim.split()
    detaylar = []

    # Sondan başlayarak, kısaltmaları ayır
    while kelimeler and kelimeler[-1] in kisaltmalar:
        detaylar.append(kelimeler.pop())  # Son kelimeyi detaylara ekle

    return " ".join(kelimeler), " ".join(reversed(detaylar)) if detaylar else None

# 📌 Yeni sütunları ekleyelim
df[["At İsmi", "Detaylar"]] = df["At İsmi"].apply(lambda x: pd.Series(ayir_at_isim(x)))

# 📌 Sonuçları görelim
df
```

```python
# Ön temizlik (emin olmak için)
df_valid = df_merged.dropna(subset=["At No", "Antrenör Adı", "Sahip Adı"])
df_valid["At No"] = pd.to_numeric(df_valid["At No"], errors="coerce")
df_valid["Kazandı mı"] = df_valid["At No"] == 1

# ANRENÖR SKORU
antrenor_scores = df_valid.groupby("Antrenör Adı").agg({
    "At No": ["mean", "count"],
    "Kazandı mı": "sum"
}).reset_index()

antrenor_scores.columns = ["Antrenör Adı", "Ortalama Sıra", "Toplam Yarış", "1.lik Sayısı"]
antrenor_scores["Antrenör_Skor"] = antrenor_scores["1.lik Sayısı"] * 10 - antrenor_scores["Ortalama Sıra"]
antrenor_scores = antrenor_scores.sort_values(by="Antrenör_Skor", ascending=False)

# SAHİP (OWNER) SKORU
sahip_scores = df_valid.groupby("Sahip Adı").agg({
    "At No": ["mean", "count"],
    "Kazandı mı": "sum"
}).reset_index()

sahip_scores.columns = ["Sahip Adı", "Ortalama Sıra", "Toplam Yarış", "1.lik Sayısı"]
sahip_scores["Sahip_Skor"] = sahip_scores["1.lik Sayısı"] * 10 - sahip_scores["Ortalama Sıra"]
sahip_scores = sahip_scores.sort_values(by="Sahip_Skor", ascending=False)
```

```python
import re

def parse_kilo(kilo_str):
    if pd.isna(kilo_str):
        return None

    # Virgülü noktaya çevir (Türk formatı → İngiliz formatı)
    kilo_str = kilo_str.replace(",", ".")

    # Rakam ve artı kısmını ayır (örn: "58.5 +0.20")
    parts = re.findall(r"[\d.]+", kilo_str)

    try:
        numbers = [float(p) for p in parts]
        return sum(numbers)
    except:
        return None

# Uygula
df_merged["Kilo"] = df_merged["Kilo"].astype(str).apply(parse_kilo)

# Kontrol
print(df_merged["Kilo"].head())
```

```python
import re

def derece_to_milliseconds(derece_str):
    if pd.isna(derece_str):
        return None

    # Önce ":" varsa "." ile değiştir
    derece_str = derece_str.replace(":", ".")

    # Derece örneği: "1.28.36" veya "0.59.72"
    parts = derece_str.split(".")

    try:
        # Dakika, saniye, salise?
        if len(parts) == 3:
            dakika = int(parts[0])
            saniye = int(parts[1])
            salise = int(parts[2])
        elif len(parts) == 2:
            dakika = 0
            saniye = int(parts[0])
            salise = int(parts[1])
        else:
            # Sadece saniye gibi düşünebiliriz
            dakika = 0
            saniye = 0
            salise = int(parts[0])

        # Saliseye çevir: 1 dakika = 60000 ms, 1 saniye = 1000 ms, 1 salise = 10 ms (tahmini)
        toplam_ms = dakika * 60000 + saniye * 1000 + salise * 10
        return toplam_ms
    except:
        return None

# Uygula
df_merged["Derece_ms"] = df_merged["Derece"].astype(str).apply(derece_to_milliseconds)

print(df_merged[["Derece", "Derece_ms"]].head(10))
```

```python
def doldur_st(grup):
    # Mevcut start numaraları (St) → dolu olanlar
    mevcut_st = set(grup["St"].dropna().astype(int))

    # Tüm mümkün numaralar (1’den toplam yarışçı sayısına kadar)
    tum_st = set(range(1, len(grup) + 1))

    # Eksik olanlar (kullanılmamış start numaraları)
    eksik_st = list(tum_st - mevcut_st)

    # Eksik `St` olan indeksler
    eksik_indeksler = grup[grup["St"].isna()].index

    # Eşleştir
    for idx, st_no in zip(eksik_indeksler, sorted(eksik_st)):
        grup.at[idx, "St"] = st_no

    return grup

# Uygula: Yarış_Id'ye göre gruplandırarak her grupta eksikleri doldur
df_merged = df_merged.groupby("Yarış_Id").apply(doldur_st)
```
