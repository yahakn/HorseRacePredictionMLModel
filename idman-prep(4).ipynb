```python
import pandas as pd

# Süreyi saniyeye çeviren fonksiyon (dakika.saniye.salise formatı için)
def sureyi_saniyeye_cevir(sure_str):
    if pd.isna(sure_str):
        return None
    parts = sure_str.split('.')
    try:
        if len(parts) == 3:
            dakika, saniye, salise = parts
            dakika = int(dakika)
            saniye = int(saniye)
            salise = int(salise)
            toplam_saniye = dakika * 60 + saniye + salise / 100
            return toplam_saniye
        elif len(parts) == 2:  # dakika.saniye formatı
            dakika, saniye = parts
            dakika = int(dakika)
            saniye = float(saniye)
            toplam_saniye = dakika * 60 + saniye
            return toplam_saniye
        else:
            # Eğer format farklıysa direkt float dönüşmeyi dene
            return float(sure_str)
    except:
        return None

# Süre sütunları listesi
sure_sutunlari = ['1400m Süre', '1200m Süre', '1000m Süre', '800m Süre', '600m Süre', '400m Süre', '200m Süre']

# Yeni sütunları saniye cinsinden oluştur
for sutun in sure_sutunlari:
    yeni_sutun_adi = sutun + ' (saniye)'
    df[yeni_sutun_adi] = df[sutun].apply(sureyi_saniyeye_cevir)

# Örnek: İlk 5 satırı göster
print(df[[sutun for sutun in df.columns if 'saniye' in sutun]].head())
```

```python
dosya_yolu = "/content/drive/My Drive/Adsız klasör/tarih/at_yarisi_numerik.csv"
import pandas as pd
yarışlar = pd.read_csv(dosya_yolu)
yarışlar.head()  # Veriyi görüntüle
```

```python
# Diyelim ki at_id tablosunda 'At İsmi' ve 'At_Id' sütunları var
# İdman setinin adı idmanlar ve at_id tablosu da at_id

# Önce idmanlardaki at isimlerini id'ye dönüştürmek için merge yapalım
idmanlar = idmanlar.merge(at_id[['At İsmi', 'At_Id']], left_on='At Adı', right_on='At İsmi', how='left')

# İdmanlarda At_Id olmayanları çıkar (isim eşleşmeyenler)
idmanlar = idmanlar[idmanlar['At_Id'].notna()]

# Artık 'At Adı' sütununu kaldırabiliriz veya istersen kaldırma
idmanlar = idmanlar.drop(columns=['At Adı', 'At İsmi'])

# 'At_Id' sütununu isim yerine kullan
idmanlar = idmanlar.rename(columns={'At_Id': 'At No'})

# İstersen index resetle
idmanlar = idmanlar.reset_index(drop=True)
```

```python
idmanlar = idmanlar.rename(columns={'At No':'At_Id'})
```

```python
yarışlar['Tarih'] = pd.to_datetime(yarışlar['Tarih'], dayfirst=True, errors='coerce')
idmanlar['İdman Tarihi'] = pd.to_datetime(idmanlar['İdman Tarihi'], dayfirst=True, errors='coerce')
```

```python
# Tarihleri datetime'a çevir
yarışlar['Tarih'] = pd.to_datetime(yarışlar['Tarih'])
idmanlar['İdman Tarihi'] = pd.to_datetime(idmanlar['İdman Tarihi'])

# At ID'ye göre birleştir (yarışlara idmanları ekle)
birlesik = pd.merge(yarışlar[['At_Id', 'Tarih']], idmanlar, on='At_Id', how='left')

# Yalnızca yarıştan önceki idmanlar
birlesik = birlesik[birlesik['İdman Tarihi'] < birlesik['Tarih']]
```

```python
birlesik
```

```python
# İdmanları zamana göre sırala ve her yarış için son 5 taneyi numaralandır
birlesik = birlesik.sort_values(['At_Id', 'Tarih', 'İdman Tarihi'], ascending=[True, True, False])
birlesik['idman_sırası'] = birlesik.groupby(['At_Id', 'Tarih']).cumcount() + 1

# Sadece son 5 idmanı al
birlesik = birlesik[birlesik['idman_sırası'] <= 5]
```

```python
# Süre sütunları
sure_sutunlar = [
    '1400m Süre (saniye)', '1200m Süre (saniye)', '1000m Süre (saniye)',
    '800m Süre (saniye)', '600m Süre (saniye)', '400m Süre (saniye)', '200m Süre (saniye)'
]

# Sayısal tipte olmayanları float'a çevir
for col in sure_sutunlar:
    birlesik[col] = pd.to_numeric(birlesik[col], errors='coerce')

# Ortalama süreleri hesapla
ortalama_sureler = birlesik.groupby(['At_Id', 'Tarih'])[sure_sutunlar].mean().reset_index()

# Yarış setine geri merge et
yarışlar = pd.merge(yarışlar, ortalama_sureler, on=['At_Id', 'Tarih'], how='left', suffixes=('', '_idman_ort'))
```

```python
yarışlar.info()
```

```python
# Tüm sütunları kontrol et ve tamamen boş (non-null = 0) olanları sil
yarışlar = yarışlar.dropna(axis=1, how='all')
```

```python
yarışlar
```

```python
at_jokey_uyumu = (
    yarışlar.groupby(["At_Id", "Jokey_Skor"])["At No"]
    .mean()
    .reset_index()
    .rename(columns={"At No": "At_Jokey_Uyumu"})
)

yarışlar = yarışlar.merge(at_jokey_uyumu, on=["At_Id", "Jokey_Skor"], how="left")
```

```python
# Pist tipine göre At No ortalamaları
at_pist_kum = yarışlar[yarışlar["Pist_Kum"] == True].groupby("At_Id")["At No"].mean().reset_index()
at_pist_kum.rename(columns={"At No": "At_Pist_Kum_Uyumu"}, inplace=True)

at_pist_sentetik = yarışlar[yarışlar["Pist_Sentetik"] == True].groupby("At_Id")["At No"].mean().reset_index()
at_pist_sentetik.rename(columns={"At No": "At_Pist_Sentetik_Uyumu"}, inplace=True)

at_pist_cim = yarışlar[yarışlar["Pist_Çim"] == True].groupby("At_Id")["At No"].mean().reset_index()
at_pist_cim.rename(columns={"At No": "At_Pist_Çim_Uyumu"}, inplace=True)

# Merge'leme işlemleri
yarışlar = yarışlar.merge(at_pist_kum, on="At_Id", how="left")
yarışlar = yarışlar.merge(at_pist_sentetik, on="At_Id", how="left")
yarışlar = yarışlar.merge(at_pist_cim, on="At_Id", how="left")
```

```python
def mesafe_kategorisi(m):
    if m < 1400:
        return "Kısa"
    elif m <= 1900:
        return "Orta"
    else:
        return "Uzun"

yarışlar["Mesafe_Kategori"] = yarışlar["Mesafe"].apply(mesafe_kategorisi)

at_mesafe_uyumu = (
    yarışlar.groupby(["At_Id", "Mesafe_Kategori"])["At No"]
    .mean()
    .reset_index()
    .rename(columns={"At No": "At_Mesafe_Uyumu"})
)

yarışlar = yarışlar.merge(at_mesafe_uyumu, on=["At_Id", "Mesafe_Kategori"], how="left")
```

```python
# Mesafe kategorisi belirleyelim
def mesafe_kategorisi(mesafe):
    if mesafe <= 1400:
        return "Kısa"
    elif mesafe <= 1900:
        return "Orta"
    else:
        return "Uzun"

yarışlar["Mesafe_Kategori"] = yarışlar["Mesafe"].apply(mesafe_kategorisi)

# At ve mesafe kategorisine göre ortalama At No'sunu hesapla (düşük sıra = iyi performans)
ortalama_sira = yarışlar.groupby(["At_Id", "Mesafe_Kategori"])["At No"].mean().reset_index()

# En iyi mesafe kategorisini bulmak için önce At_Id bazında minimum ortalama sıra değerini bul
min_sira = ortalama_sira.groupby("At_Id")["At No"].min().reset_index().rename(columns={"At No": "Min_Ortalama_Sira"})

# İyi olan mesafe kategorisini alalım
best_mesafe = pd.merge(ortalama_sira, min_sira, on="At_Id")
best_mesafe = best_mesafe[best_mesafe["At No"] == best_mesafe["Min_Ortalama_Sira"]]

# Atın tercih ettiği mesafe kategorisi
best_mesafe = best_mesafe[["At_Id", "Mesafe_Kategori"]].rename(columns={"Mesafe_Kategori": "Mesafe_Tercihi"})

# Mesafe tercihini df'ye ekleyelim
yarışlar = yarışlar.merge(best_mesafe, on="At_Id", how="left")
```

```python
yarışlar = pd.get_dummies(yarışlar, columns=['Mesafe_Tercihi'], prefix='Mesafe')
```

```python
yarışlar = yarışlar.drop(columns=['Mesafe_Kategori', 'At_Mesafe_Uyumu', 'Önceki_Tarih'])
```

```python
# Her yarışta kaç at olduğunu hesapla
at_sayilari = yarışlar.groupby("Yarış_Id")["At No"].count()

# Ortalama at sayısı
ortalama_at_sayisi = at_sayilari.mean()

print(f"Her yarışta ortalama at sayısı: {ortalama_at_sayisi:.2f}")
```

```python
# 1. Sıralama
yarışlar = yarışlar.sort_values(by=["Yarış_Id", "At No"]).reset_index(drop=True)

# 2. Her yarış için tarih bilgisini al (Tarih bilgisi olan satırdan)
tarih_map = yarışlar.groupby("Yarış_Id")["Tarih"].apply(lambda x: x.dropna().iloc[0] if x.dropna().any() else None)

# 3. Tarihi olmayan satırlara o yarışın tarihini ekle
yarışlar["Tarih"] = yarışlar.apply(lambda row: tarih_map[row["Yarış_Id"]] if pd.isna(row["Tarih"]) else row["Tarih"], axis=1)
```
