```python
def derece_to_saniye(derece_str):
    try:
        parts = derece_str.split(".")
        if len(parts) == 3:
            dakika, saniye, salise = parts
            return int(dakika) * 60 + int(saniye) + int(salise) / 100
        elif ":" in derece_str:
            dakika, saniye_salise = derece_str.split(":")
            saniye, salise = saniye_salise.split(".")
            return int(dakika) * 60 + int(saniye) + int(salise) / 100
        else:
            return float(derece_str)
    except:
        return None
```

```python
# 📌 Önce virgülleri noktaya çeviriyoruz
df["Kilo"] = df["Kilo"].astype(str).str.replace(",", ".", regex=False)

# 📌 Sonra + işareti olanları toplayıp, olmayanları doğrudan floata çeviriyoruz
def kilo_duzelt(kilo):
    if pd.isna(kilo):
        return None
    if "+" in kilo:
        ana, arti = kilo.split("+")
        return float(ana.strip()) + float(arti.strip())
    try:
        return float(kilo)
    except ValueError:
        return None

df["Kilo"] = df["Kilo"].apply(kilo_duzelt)
```

```python
import re

def temizle_yas(x):
    if pd.isna(x):
        return x  # NaN ise dokunma
    x = str(x).strip()  # Baş ve son boşlukları temizle
    match = re.search(r'\d+', x)  # İlk geçen rakamı bul
    if match:
        return int(match.group())  # Bulduysa integer yap
    else:
        return None  # Sayı bulamazsa None yap

# 📌 Uygulayalım
df["Yaş"] = df["Yaş"].apply(temizle_yas)
```

```python
df["St"] = df["St"].astype(str).str.extract(r"(\d+)")  # Sayıları ayıkla
df["St"] = df["St"].astype(int)  # Integer'a çevir
```

```python
df["H"] = df["H"].astype(str).str.extract(r"(\d+)")  # Sayıları ayıkla
df["H"] = df["H"].astype(int)  # Integer'a çevir
```

```python
def derece_to_saniye(d):
    try:
        parts = d.split(".")
        if len(parts) == 3:
            dakika, saniye, salise = parts
            return int(dakika) * 60 + int(saniye) + int(salise) / 100
        elif len(parts) == 2:
            dakika, saniye = parts
            return int(dakika) * 60 + float(saniye)
        else:
            return float(d)
    except:
        return None

df["Derece_saniye"] = df["Derece"].apply(derece_to_saniye)
```

```python
df["Derece_saniye"] = df["Derece_saniye"].astype(float)  # Integer'a çevir
```

```python
df = df.drop(columns=["Ganyan"])
```

```python
df["Sıklet"] = df["Sıklet"].astype(str).str.extract(r"(\d+)")  # Sayıları ayıkla
df["Sıklet"] = df["Sıklet"].astype(int)  # Integer'a çevir
```

```python
df = pd.get_dummies(df, columns=["Pist"], prefix="Pist")
```

```python
# Baba için skor hesaplama
baba_grup = df.groupby("Orijin(Baba)")

baba_skor = baba_grup.apply(lambda x: (
    0.5 * (x["At No"] == 1).mean() +           # Birincilik oranı ağırlığı
    0.3 * (x["At No"] <= 3).mean() -           # İlk 3 oranı ağırlığı
    0.2 * (x["At No"].mean() / df["At No"].max())  # Ortalama sıralama normalize edilip ağırlıklandı
)).reset_index(name="Baba_Skor")

# Anne için skor hesaplama
anne_grup = df.groupby("Orijin(Anne)")

anne_skor = anne_grup.apply(lambda x: (
    0.5 * (x["At No"] == 1).mean() +
    0.3 * (x["At No"] <= 3).mean() -
    0.2 * (x["At No"].mean() / df["At No"].max())
)).reset_index(name="Anne_Skor")

# Skorları df'ye ekle
df = df.merge(baba_skor, on="Orijin(Baba)", how="left")
df = df.merge(anne_skor, on="Orijin(Anne)", how="left")

```

```python
def hesapla_skor(grup):
    return grup.apply(lambda x: (
        0.5 * (x["At No"] == 1).mean() +            # Birincilik oranı
        0.3 * (x["At No"] <= 3).mean() -            # İlk 3 oranı
        0.2 * (x["At No"].mean() / df["At No"].max())  # Ortalama pozisyon (normalize)
    ))

# At sahibi
sahip_grup = df.groupby("Sahip Adı")
sahip_skor = hesapla_skor(sahip_grup).reset_index(name="Sahip_Skor")

# Antrenör
antrenor_grup = df.groupby("Antrenör Adı")
antrenor_skor = hesapla_skor(antrenor_grup).reset_index(name="Antrenor_Skor")

# Jokey
jokey_grup = df.groupby("Jokey Adı")
jokey_skor = hesapla_skor(jokey_grup).reset_index(name="Jokey_Skor")

# Skorları df'ye ekleyelim
df = df.merge(sahip_skor, on="Sahip Adı", how="left")
df = df.merge(antrenor_skor, on="Antrenör Adı", how="left")
df = df.merge(jokey_skor, on="Jokey Adı", how="left")

# İstersen isim sütunlarını düşüp sadece skorları bırakabilirsin:
#df = df.drop(columns=["Sahip Adı", "Antrenör Adı", "Jokey Adı"])
```
